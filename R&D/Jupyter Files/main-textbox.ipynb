```python
import shutil
import os
import re
from shutil import copyfile
```

```python
def main_crop(direc,main_folder_fname,text_box_fname):
    
    # main_folder path contains label_images & text_box folders
    #direc=r'C:\\Users\\anilkumar.maddikunta\\Desktop\\prescreening resumes' 
    
    #img_path='\\main'  # label images folder --> '\\folder_name'
    img_path=main_folder_fname
    main_img_path=direc+img_path
    main_file_names = os.listdir(main_img_path)
    all_names=[]
    for file_name in main_file_names: 
        name=file_name[:-6]
        if name not in all_names:
            all_names.append(name)
    
    textBox_path='\\text_box'  # text_boxs folder --> '\\folder_name'
    textBox_path=text_box_fname
    textBox_img_path=direc+textBox_path
    textBox_file_names = os.listdir(textBox_img_path)
            
    
    new_direc_path = direc+'\\main-textBox'
    try:
        os.mkdir(new_direc_path)
    except OSError:
        pass
    else:
        pass   
    for name in all_names:
        name_path='\\'+name
        new_name_path = new_direc_path + name_path
        try:
            os.mkdir(new_name_path)
        except OSError:
            pass
        else:
            pass
        
        os.chdir(new_name_path)
        
        new_name_main_path=new_name_path+'\\main'
        try:
            os.mkdir(new_name_main_path)
        except OSError:
            pass
        else:
            pass
        for file_name in main_file_names:
            name_=file_name[:-6]
            if name==name_:
                shutil.copy(os.path.join(main_img_path, file_name), new_name_main_path)
                
                
        new_name_text_box_path=new_name_path+'\\text_box'
        try:
            os.mkdir(new_name_text_box_path)
        except OSError:
            pass
        else:
            pass                
        
        for file_name in textBox_file_names:
            name_=file_name[:-6]
            name2_=file_name[:-7]
            name3_=file_name[:-8]
            if name==name_:
                shutil.copy(os.path.join(textBox_img_path, file_name), new_name_text_box_path)
            if name==name2_:
                shutil.copy(os.path.join(textBox_img_path, file_name), new_name_text_box_path)
                
            #if name==name3_:
                #shutil.copy(os.path.join(textBox_img_path, file_name), new_name_text_box_path)

    
```
