```python
def readresume(path):
    with open(path, "r",encoding='utf-8') as fp:
        text = ''.join(fp.readlines())
        return text

def resume_to_csv(rootpath):
    lst = []
    for f in listdir(rootpath):
        inner_path = join(rootpath, f)
        lst.extend([(readresume(join(inner_path, ip)),f, ip[:-4]) for ip in listdir(inner_path)])
    df = pd.DataFrame(lst, columns =["resume", 'role', 'name'])
    df.to_csv(f"{rootpath}\\resume_data_csv.csv", index = False)
    
resume_to_csv(mypath)   
```

 ### Similarity Score for skills in Resume and JD using TF_IDF

```python
tfidf = TfidfVectorizer(sublinear_tf= True, #use a logarithmic form for frequency
                       min_df = 4, #minimum numbers of documents a word must be present in to be kept
                       norm= 'l2', #ensure all our feature vectors have a euclidian norm of 1
                       ngram_range= (1,2), #to indicate that we want to consider both unigrams and bigrams.
                       stop_words ='english')

features_jd = tfidf.fit_transform(df_jd['jd_skills'])

features_resume = tfidf.transform(dff['skills'])

cosine_sim = cosine_similarity(features_jd, features_resume)
```

```python
indices = pd.Series(df_jd.index, index=df_jd['jd_skills'])
```

```python
def score_generation(jd_skills, cosine_sim=cosine_sim):

    idx = indices[jd_skills]

    sim_scores   = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:50]  

    resume_indices = [i[0] for i in sim_scores]
    resume_scores = [i[1] for i in sim_scores]
    
    final_resume=dff['skills'].iloc[resume_indices]
    
    final_name = dff['name'].iloc[resume_indices]
    
    final_role = dff['role'].iloc[resume_indices]
    
    final_scores = pd.DataFrame()
    
    final_scores['resume'] = final_resume
    
    final_scores['scores'] = resume_scores
    
    final_scores['name'] = final_name
    
    final_scores['role'] = final_role   
    
    final_scores.to_csv(r'E:\work\op_work\O-Staff\scg\SSG\result\tf_idf\final_scores_ux.csv')
    
    return final_scores
```

```python
jd = df_jd['jd_skills'][5]
final_result = score_generation(jd)
# print(final_result)
```

```python
jd = df_jd['jd_skills'][0]
final_result = score_generation(jd)
print(final_result)
```
